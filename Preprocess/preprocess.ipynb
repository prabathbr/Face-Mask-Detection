```python
# Parameters

base_directory = "e:/image_classify/" # base directory
original_images ="dataset_part1" # original dataset directory
temp_base = "dataset_cleaned_part1" # temporary dataset directory
classify_names = ["type1_male", "type2_male", "type3_male", "type4_male","type1_female", "type2_female", "type3_female", "type4_female"] # class names to be classified into
```

```python
# Change to base directory

import os
import shutil


os.chdir(base_directory)
print("base directory : " + os.getcwd())
```

```python
from PIL import Image
import imghdr

def check_image(img_filename):
    image_type = imghdr.what(img_filename)
    if image_type == 'jpeg':
        # Reference : https://stackoverflow.com/a/53470882/13492054
         try:
            im = Image.open(img_filename)
            im.verify()
            im.close()
            im = Image.open(img_filename)
            im.transpose(Image.FLIP_LEFT_RIGHT)
            im.close()
            return True
         except:
            print("File Error: " + img_filename)
            return False

 
```

```python
# Creating directories for the cleaned dataset

try:
    shutil.rmtree(base_directory + temp_base,ignore_errors=True)
    os.makedirs(base_directory + temp_base, exist_ok=True)
    
    for names in classify_names:
        os.makedirs(base_directory + temp_base + "/" + names, exist_ok=True)
    
except:
    print("Error creating directories for the cleaned dataset")       
```

```python
# Query all file names in original_images

file_list = os.listdir(base_directory+original_images)
print("Total number of files: " + str(len(file_list)))
```

```python
# Copying files to cleaned dataset directories

file_number = 0
success_files = []
error_files = []


for file in file_list:
    file_info = file.split("_")
    
    mask_type = file_info[1]
    gender = file_info[3].lower()
    file_number += 1
    
    classify_dir = "type"+ str(mask_type) + "_" + gender
    dst_file_name = str(file_number) + "_" + classify_dir + ".jpg"
        
    src = base_directory + original_images + "/" + file
    dst = base_directory + temp_base + "/"  + classify_dir + "/" + dst_file_name
    
    is_jpg = False
    is_jpg = check_image(src) 
    
    if ((mask_type == "1" or mask_type == "2" or mask_type == "3" or mask_type == "4") and (gender == "male" or gender == "female") and is_jpg == True):
        #print("Copying " + src + " to " + dst)
        print(str(file_number) + " out of " + str(len(file_list)))
        shutil.copy2(src, dst) 
        success_files.append(src)
    else:
        error_files.append(src)     


print("Total number of files: " + str(len(file_list)))        
print("Total number of successful files: " + str(len(success_files)))  
print("Total number of error files: " + str(len(error_files)))
```
